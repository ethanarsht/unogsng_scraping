```python
import pandas as pd
import numpy as np
import requests
import io
import json
```

```python
df_titles = pd.read_csv('combined_titles.csv').drop('Unnamed: 0', axis = 1)
```

```python
df_imdb_true = df_titles.dropna(subset = ['imdbid'])
```

```python
df_all = pd.DataFrame()
for id in df_imdb_true.nfid:
    print(id)
    import requests

    url = "https://unogsng.p.rapidapi.com/title"

    querystring = {"netflixid":str(id)}

    headers = {
        'x-rapidapi-host': "unogsng.p.rapidapi.com",
        'x-rapidapi-key': "1c41c506a8mshd5207c6b7cc3c8bp1f40d2jsn36dbb18cbd7b"
        }

    response = requests.request("GET", url, headers=headers, params=querystring)

    json_data = json.loads(response.text)['results']

    df_line = pd.DataFrame(json_data)
    df_line['nfid'] = id

    df_all = pd.concat([df_line, df_all])
```

```python
df_all_select = df_all.drop(['vtype', 'img', 'imdbid', 'title', 'synopsis', 'year', 'imdbrating', 'avgrating'], axis = 1)
```

```python
df_combined = pd.merge(
    df_titles, df_all_select, 
    how = 'left',
    on = 'nfid')
```
